# Pr√©-processamento de dados

```python
import sys
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
```

```python
def save_data(df, data_set):
    # Normalizar colunas por reescala (m√°ximo absoluto) com exce√ß√£o da classifica√ß√£o
    scaler = MinMaxScaler(feature_range = (0, 1))

    # for col in df.drop(['CLASS', 'PROTOCOL', 'FLAGS', 'SRC_IP', 'SRC_PT', 'DST_IP', 'DST_PT'], axis = 1):
    for col in df.drop(['CLASS'], axis = 1):
        display(f'{col} - normalizado por {scaler}')
        df[[col]] = scaler.fit_transform(df[[col]])
        
    # Correla√ß√£o
    f, ax = plt.subplots(figsize = (15, 15))
    sns.heatmap(df.corr(), annot = True, linewidths = .5, fmt = '.2f')
    ax.set_title(f'Matriz de correla√ß√£o ({data_set})')
    plt.savefig(f'./results/images/heat-map-{data_set}.jpg', bbox_inches = 'tight', dpi = 150)
    plt.show()

    # Analisar colunas
    for c in df.columns:
        plt.scatter(range(df.shape[0]), np.sort(df[c].values))
        plt.xlabel('index')
        plt.ylabel(c)
        plt.title(f'Distribui√ß√£o de "{c}" ({data_set})')
        sns.despine()
        plt.savefig(f'./results/images/distribuicao-{data_set}-{c}.jpg', bbox_inches = 'tight', dpi = 150)
        plt.show()

    # An√°lise de relacionamento dos atributos do conjunto de dados
    display(df.var())
    df.var().to_csv(f'./results/reports/CIDDS-001-{data_set}-variancia-pre-processed.csv')
    df.var().to_latex(f'./results/reports/CIDDS-001-{data_set}-variancia-pre-processed.tex')

    # Estat√≠sticas p√≥s processamento
    display(df.describe(include = 'all'))

    # Distribui√ß√£o dos atributos
    for attribute in df[['PROTOCOL', 'FLAGS', 'CLASS']]:
        plt.clf()
        plt.figure(figsize = (12, 8))
        mpl.rcParams.update({
            'axes.titlesize': '18',
            'xtick.labelsize': '14',
            'ytick.labelsize': '14'
        })
        plt.title(f'Distribui√ß√£o de {attribute} ({data_set})')
        df[attribute].value_counts().apply(np.log).plot(kind = 'barh')
        plt.savefig(f'./results/images/distribuicao-{data_set}-{attribute}-valores.jpg', bbox_inches = 'tight', dpi = 150)
        plt.show()

    # Estat√≠sticas p√≥s processamento
    df_describe = df.describe(include = 'all')
    df_describe.fillna(0, inplace = True)
    df_describe.to_csv(f'./results/reports/CIDDS-001-{data_set}-stats-after-pre-processed.csv')
    df_describe.to_latex(f'./results/reports/CIDDS-001-{data_set}-stats-after-pre-processed.tex')
        
    # Salvar os dados pr√©-processados
    data = df.drop('CLASS', axis = 1)
    label = pd.DataFrame(df['CLASS'])
    display(data.shape)
    display(label.shape)

    X_train, X_test, y_train, y_test = train_test_split(data, label, test_size = 0.2)
    
    display('Dados de treinamento')
    display(X_train.shape)   
    display(y_train.shape)
    
    display('Dados de teste')
    display(X_test.shape)   
    display(y_test.shape)

    X_train.to_csv(f'./results/data/CIDDS-001-{data_set}-pre-processed-train-data.csv', index = False)
    y_train.to_csv(f'./results/data/CIDDS-001-{data_set}-pre-processed-train-class.csv', index = False)
    X_test.to_csv(f'./results/data/CIDDS-001-{data_set}-pre-processed-test-data.csv', index = False)
    y_test.to_csv(f'./results/data/CIDDS-001-{data_set}-pre-processed-test-class.csv', index = False)
    df.to_csv(f'./results/data/CIDDS-001-{data_set}-pre-processed.csv', index = False)
    
random_seed = 42
random_state = np.random.RandomState(random_seed)

# contabilizar classe
stats_class = pd.DataFrame(columns = [])
stats_class_normal = pd.DataFrame(columns = [])

# Arquivo de dados
data_sets = [
    ('internal-week1'),
#     ('internal-week2'),
#     ('internal-week3'), # n√£o tem classes variadas (s√≥ 1=normal)
#     ('internal-week4'), # n√£o tem classes variadas (s√≥ 1=normal)
    
#     ('external-week1'),
#     ('external-week2'),    
#     ('external-week3'),
#     ('external-week4'),
]

row_index = 0
for data_set in data_sets:
    display(f'{data_set}')
        
    df = pd.read_csv(f'./data/CIDDS-001-{data_set}.csv', dtype = {
            'Date first seen': 'object',
            'Duration': 'float64',
            'Proto': 'object',
            'Src IP Addr': 'object',
            'Src Pt': 'int64',
            'Dst IP Addr': 'object',
            'Dst Pt': 'float64',
            'Packets': 'int64',
            'Bytes': 'object',
            'Flows': 'int64',
            'Flags': 'object',
            'Tos': 'int64',
            'class': 'object',
            'attackType': 'object',
            'attackID': 'object',
            'attackDescription': 'object',
        }, parse_dates = ['Date first seen'])
    df = df.rename(columns = {
        'Duration': 'DURATION',
        'Proto': 'PROTOCOL',
        'Packets': 'PACKETS',
        'Bytes': 'BYTES',
        'Flags': 'FLAGS',
        'class': 'CLASS',
    })

    # Eliminar atributos que n√£o ser√£o utilizados (valores n√£o s√£o consistentes para todos conjuntos de dados)
    df = df.drop(['Date first seen', 'Src IP Addr', 'Src Pt', 'Dst IP Addr', 'Dst Pt', 'Flows', 'Tos', 'attackType', 'attackID', 'attackDescription'], axis = 1)

    # Estat√≠sticas e formatos iniciais
    display(f'Shape: {df.shape}')
    df_describe = df.describe(include = 'all')
    df_describe.fillna(0, inplace = True)
    display(df_describe)
    df_describe.to_csv(f'./results/reports/CIDDS-001-{data_set}-stats-before-pre-processed.csv')
    df_describe.to_latex(f'./results/reports/CIDDS-001-{data_set}-stats-before-pre-processed.tex')
    display(df.dtypes)

    for name, row in df['CLASS'].value_counts().to_frame().iterrows():
        stats_class.loc[row_index, 'CLASS'] = name
        stats_class.loc[row_index, 'QTD'] = row['CLASS'].astype(int)
        
        row_index += 1

    # Subsampling excluindo 'v√≠timas'
    display('Subsampling')
    display(df['CLASS'].value_counts())

    df = df[df['CLASS'] != 'victim']
    df = df.groupby('CLASS').apply(lambda x: x[x['CLASS'] != 'victim'].sample(frac = 0.004, replace = True, random_state = random_state))

    display(df['CLASS'].value_counts())
    display(df.shape)

    # Converter coluna 'Bytes' para valor num√©rico
    def toNumber(df):
        num = df['BYTES']
        # Converte M para 
        if "M" in str(num):
            num = float("".join(re.findall("[0-9.0-9]",df['BYTES'])))*1000000
        num = int(num)
        return num

    df['BYTES'] = df.apply(toNumber, axis = 1)
    df['BYTES'] = df['BYTES'].apply(np.int64)

    # Fun√ß√£o de convers√£o para inteiro
    def toInteger(df, col):
        num1 = df[col]
        num2 = int(df[col])
        if num2 > 0 and num1 % num2 > 0:
            num3 = int(num1 * 1000)
        else:
            num3 = int(num1)
        return int(num3)

    # Converter 'Duration' para inteiro
    def toInt2(df):
        return toInteger(df, 'DURATION')

    df['DURATION'] = df.apply(toInt2, axis = 1)

    # Traduzir os valores de 'Flags'
    FLAGS_LIST = 'CEUAPRSF'

    def toFlagValue(df):
        hex_value = df['FLAGS']
        if '0x' in hex_value:
            binary_value = bin(int(hex_value, 16))[2:].zfill(8).replace('0', '.')
            flag_value = ''.join([FLAGS_LIST[i:i+1] if c  ==  '1' else c for i, c in enumerate(binary_value)])
        else:
            flag_value = hex_value
            
        return flag_value

    df['FLAGS'] = df.apply(toFlagValue, axis = 1)

    # Converter os endere√ßos IP, protocolo e flags para c√≥digos categ√≥ricos
    oe = OrdinalEncoder(dtype = int)
    df['PROTOCOL'] = oe.fit_transform(df[['PROTOCOL']])
    df['FLAGS'] = oe.fit_transform(df[['FLAGS']])

    for name, row in df['CLASS'].value_counts().to_frame().iterrows():
        stats_class_normal.loc[row_index, 'CLASS'] = name
        stats_class_normal.loc[row_index, 'QTD'] = row['CLASS'].astype(int)
        
        row_index += 1
    
    # Converter "CLASS" para c√≥digo categ√≥rio pr√©-determinado
    # Primeiro o diferente sen√£o o besta pega todo mundo!!
    df.replace({'CLASS': {r'\b(?!normal\b)\w+': '1'}}, regex = True, inplace = True)
    df.replace({'CLASS': {'normal': '0'}}, inplace = True)
    df['CLASS'] = df['CLASS'].astype(int)

    # Moda
    mode = df.mode()
    mode.fillna(0, inplace = True)
    mode.to_csv(f'./results/reports/CIDDS-001-{data_set}-stats-mode.csv', index = False)
    mode.to_latex(f'./results/reports/CIDDS-001-{data_set}-stats-mode.tex', index = False)

    # Curtose enviesada
    kurt = df.kurt().reset_index()
    kurt.rename(columns = {'index': 'FEATURE', 0: 'KURTOSIS'}, inplace = True)
    display(kurt)
    kurt.fillna(0, inplace = True)
    kurt.to_csv(f'./results/reports/CIDDS-001-{data_set}-stats-kurt.csv', index = False)
    kurt.to_latex(f'./results/reports/CIDDS-001-{data_set}-stats-kurt.tex', index = False)

    # Obliquidade
    skew = df.skew().reset_index()
    skew.rename(columns = {'index': 'FEATURE', 0: 'SKEW'}, inplace = True)
    display(skew)
    skew.fillna(0, inplace = True)
    skew.to_csv(f'./results/reports/CIDDS-001-{data_set}-stats-skew.csv', index = False)
    skew.to_latex(f'./results/reports/CIDDS-001-{data_set}-stats-skew.tex', index = False)
    
    # Balanceamento das classes
    save_data(df, f'{data_set}-umbalanced')

    display(f'Balanceando {data_set}')
    # Separar as classes
    df_min = df[df.CLASS == 0] # Normal √© minoria
    df_maj = df[df.CLASS == 1] # Suspeito √© maioria

    # Upsample classe minorit√°ria
    df_min_up = resample(df_min, replace = True, n_samples = len(df_maj), random_state = random_state)
    df = pd.concat([df_maj, df_min_up])

    display(df['CLASS'].value_counts())

    save_data(df, f'{data_set}-upsample')

    # Downsample classe maiorit√°ria
    df_maj_down = resample(df_maj, replace = True, n_samples = len(df_min), random_state = random_state)
    df = pd.concat([df_maj_down, df_min])

    save_data(df, f'{data_set}-downsample')

stats_class['CLASS'] = stats_class['CLASS'].str.upper()
stats_class.groupby(['CLASS']).sum().to_csv(f'./results/reports/CIDDS-001-{data_set}-groups-before-pre-processed.csv')
stats_class.groupby(['CLASS']).sum().to_latex(f'./results/reports/CIDDS-001-{data_set}-groups-before-pre-processed.tex')
display(stats_class.groupby(['CLASS']).sum())

stats_class_normal['CLASS'] = stats_class_normal['CLASS'].str.upper()
stats_class_normal.groupby(['CLASS']).sum().to_csv(f'./results/reports/CIDDS-001-{data_set}-groups-after-pre-processed.csv')
stats_class_normal.groupby(['CLASS']).sum().to_latex(f'./results/reports/CIDDS-001-{data_set}-groups-after-pre-processed.tex')
display(stats_class_normal.groupby(['CLASS']).sum())
```
